Project One(1) Traveling Ethiopia search Problem Aklilu Solomon

```python
from collections import deque

graph = {
    "Addis Ababa": ["Adama", "Ambo", "Debre Birhan"],
    "Adama": ["Assella", "Matahara"],
    "Ambo": ["Nekemte"],
    "Debre Birhan": ["Debre Sina"],
    "Debre Sina": ["Kemise"],
    "Kemise": ["Dessie"],
    "Dessie": ["Woldia"],
    "Woldia": ["Lalibela"],
    "Assella": ["Dodolla"],
    "Dodolla": ["Bale"],
    "Bale": ["Goba"]
}


class GraphSearch:
    def __init__(self, graph, initial_state, goal_state, strategy="BFS"):
        """
        graph: dict -> adjacency list {state: [neighbors]}
        initial_state: starting node
        goal_state: target node
        strategy: "BFS" or "DFS"
        """
        self.graph = graph
        self.initial = initial_state
        self.goal = goal_state
        self.strategy = strategy.upper()

    def search(self):
        if self.strategy == "BFS":
            return self._bfs()
        elif self.strategy == "DFS":
            return self._dfs()
        else:
            raise ValueError("Strategy must be 'BFS' or 'DFS'")

    def _bfs(self):
        frontier = deque([[self.initial]])  # queue of paths
        visited = set()

        while frontier:
            path = frontier.popleft()
            state = path[-1]

            if state == self.goal:
                return path

            if state not in visited:
                visited.add(state)
                for neighbor in self.graph.get(state, []):
                    new_path = path + [neighbor]
                    frontier.append(new_path)

        return None  # no solution found

    def _dfs(self):
        frontier = [[self.initial]]  # stack of paths
        visited = set()

        while frontier:
            path = frontier.pop()
            state = path[-1]

            if state == self.goal:
                return path

            if state not in visited:
                visited.add(state)
                for neighbor in reversed(self.graph.get(state, [])):
                    new_path = path + [neighbor]
                    frontier.append(new_path)

        return None  # no solution found


# Breadth-First Search
bfs_search = GraphSearch(graph, "Addis Ababa", "Lalibela", "BFS")
print("BFS Path:", bfs_search.search())

# Depth-First Search
dfs_search = GraphSearch(graph, "Addis Ababa", "Lalibela", "DFS")
print("DFS Path:", dfs_search.search())
```
