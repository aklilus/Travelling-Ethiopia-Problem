***Project 4 Traveling Ethiopia Search Problem as shown in Figure 4***

```python
class EthiopiaCoffeeGame:
    def __init__(self):
        # Graph structure
        self.map = {
            "Addis Ababa": ["Ambo", "Adama"],
            "Ambo": ["Gedo", "Nekemte"],
            "Adama": ["Mojo", "Dire Dawa"],
            "Gedo": ["Shambu", "Fincha"],
            "Nekemte":["Gimbi","Limu"],
            "Buta Jirra": ["Worabe", "Wolkite"],
            "Worabe": ["Hossana", "Durame"],
            "Wolkite": ["Bench Naji", "Tepi"],
            "Mojo": ["Kaffa", "Dilla"],
            "Dire Dawa": ["Harar", "Chiro"]
        }

        # Terminal utilities (coffee quality)
        self.utilities = {
            "Shambu": 4,
            "Fincha": 5,
            "Hossana": 6,
            "Durame": 5,
            "Gimbi":8,
            "Limu":8,
            "Bench Naji": 5,
            "Tepi": 6,
            "Kaffa": 7,
            "Dilla": 9,
            "Harar": 10,
            "Chiro": 6
        }

    def get_possible_moves(self, state):
        return self.map.get(state, [])

    def make_move(self, state, move, player):
        return move

    def is_terminal(self, state):
        return state in self.utilities

    def evaluate(self, state):
        return self.utilities.get(state, 0)
class Minimax:
    def __init__(self, game):
        self.game = game

    def minimax(self, state, is_max):
        if self.game.is_terminal(state):
            return self.game.evaluate(state)

        if is_max:  # Agent (MAX)
            best_value = float('-inf')
            for move in self.game.get_possible_moves(state):
                value = self.minimax(move, False)
                best_value = max(best_value, value)
            return best_value

        else:  # Adversary (MIN)
            best_value = float('inf')
            for move in self.game.get_possible_moves(state):
                value = self.minimax(move, True)
                best_value = min(best_value, value)
            return best_value

    def best_move(self, start_state):
        best_value = float('-inf')
        best_city = None

        for move in self.game.get_possible_moves(start_state):
            value = self.minimax(move, False)
            if value > best_value:
                best_value = value
                best_city = move

        return best_city, best_value
game = EthiopiaCoffeeGame()
agent = Minimax(game)

best_destination, coffee_quality = agent.best_move("Addis Ababa")

print("Best destination:", best_destination)
print("Guaranteed coffee quality:", coffee_quality)
```
